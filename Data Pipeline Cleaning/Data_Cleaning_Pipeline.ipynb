```python
from pandas import *
import re 
```

```python
## remove the URL 
def remove_URL(headline_text):
    url = re.compile(r'https?://\S+|www\.\S+')
    return url.sub(r'', headline_text)
Source_data['text'] = Source_data['text'].apply(remove_URL)
```

```python
Source_data
```

```python
Source_data = Source_data.loc[~((Source_data['Sentiment'] == 0))]
```

```python
Source_data
```

```python
def remove_html(headline_text):
    html=re.compile(r'<.*?>')
    return html.sub(r'',headline_text)

Source_data['text'] = Source_data['text'].apply(remove_html)
```

```python
Source_data
```

```python
# Remove Emojis 
def remove_emojis(data):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', data)
Source_data['text'] = Source_data['text'].apply(remove_emojis)
```

```python
import nltk.data
##Load language Specific .pickle file
tokenizer = nltk.data.load('tokenizers/punkt/PY3/english.pickle')
spanish_tokenizer = nltk.data.load('tokenizers/punkt/PY3/spanish.pickle')
##Different type of tokenizer
from nltk.tokenize import regexp_tokenize 
from nltk.tokenize import RegexpTokenizer 
from nltk.tokenize import WordPunctTokenizer 
from nltk.tokenize import TreebankWordTokenizer 
from nltk.tokenize import word_tokenize
##Sample initialization of token 
tokenizer = nltk.tokenize.RegexpTokenizer(r'\w+')
##Define Normalization 
normalization = None
normalization = 'stemmer'
normalization = 'lemmatizer'
##Define Vectorizer 
vectorizer = 'countvectorizer'
vectorizer = 'tfidfvectorizer'
```

```python
Source_data["text"] = Source_data["text"].apply(tokenizer.tokenize)
```

```python

def stem_tokens(tokens):
    stemmer = nltk.stem.PorterStemmer()
    tokens = [stemmer.stem(token) for token in tokens]
    return tokens
def lemmatize_tokens(tokens):
    lemmatizer = nltk.stem.WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return tokens
def normalize_tokens(normalization):
    if normalization is not None:
        if normalization == 'stemmer':
            Source_data ['text'] = Source_data['text'].apply(stem_tokens)
        elif normalization == 'lemmatizer':
            Source_data['text'] = Source_data['text'].apply(lemmatize_tokens)
        
normalize_tokens(normalization)
```

```python
import nltk
from nltk.corpus import stopwords
set(stopwords.words('english'))
```

```python
additional_stop_word = ["2018"]
def remove_stopwords(text):
    
    words_1 = [w for w in text if w not in stopwords.words('english')]
    words_2 = [w for w in words_1 if w not in additional_stop_word ]
    return words_2
def remove_numbers(text):

    word = [x for x in text if not (x.isdigit())]
    return word 
Source_data['text'] = Source_data['text'].apply(remove_stopwords)
Source_data['text'] = Source_data['text'].apply(remove_numbers)


```

```python
Source_data
```

```python
Positive_df = Source_data[Source_data["Sentiment"]== 1 ]
Negative_df = Source_data[Source_data["Sentiment"]== -1 ]
```

```python
Positive_df
```

```python
Positive_word_bank = {}

for index, row in Positive_df.iterrows():
    word_list = row['text']
    for word in word_list:
        if word in Positive_word_bank:
            Positive_word_bank[word] += 1
        else:
            Positive_word_bank[word] = 1
            
Negative_word_bank = {}

for index, row in Negative_df.iterrows():
    word_list = row['text']
    for word in word_list:
        if word in Negative_word_bank:
            Negative_word_bank[word] += 1
        else:
            Negative_word_bank[word] = 1

```

```python
### Positive World Cloud 
from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud

Positive_Cloud = WordCloud(background_color="white",width=5000,height=5000,relative_scaling=0.5,normalize_plurals=False).generate_from_frequencies(Positive_word_bank)
plt.imshow(Positive_Cloud)
```

```python
### Negative World Cloud
from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud

Negative_Cloud = WordCloud(background_color="white",width=5000,height=5000,).generate_from_frequencies(Negative_word_bank)
```
